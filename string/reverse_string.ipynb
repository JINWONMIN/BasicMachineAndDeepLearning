[Reverse String 문제 출처](https://leetcode.com/problems/reverse-string/)

Write a function that reverses a string. The input string is given as an array of characters s.

You must do this by modifying the input array in-place with 0(1) extra memory.

#### Example 1:

**Input**: s = ["h", "e", "l", "l", "o"] <br>
**Output**: ["o", "l", "l", "e", "h"]

#### Example2:

**Input**: s = ["H", "a", "n", "n", "a", "h"] <br>
**Output**: ["h", "a", "n", "n", "a", "H"]

#### Constraints:

- 1 <= s.length <= 10^5
- s[i] is a printable ascii character.

```python
from typing import List

class Solution:
    def ReverseString(self, s: List[str]) -> None:
        """
        Do not return anything, modify s in-place instead.
        """
        left, right = 0, len(s) -1
        while left < right:
            s[left], s[right] = s[right], s[left] 
            left += 1
            right -= 1
```

: 간단한 문자열 뒤집기 문제로 핵심은 in-place로 추가 메모리를 사용하지 않고 해결하라는 조건이다. 투 포인터를 이용한 스왑 방식으로 간단하게 해결할 수 있다.

left, right에 처음과 끝에 해당하는 인덱스를 부여 후 서로 스왑하는 방법으로 문자열을 뒤집을 수 있다.

```python
class Solution:
    def ReverseString(self, s: List[str]) -> None:
        s.reverse()
```

: 간단하게 파이썬 내장함수인 reverse()를 이용해 뒤집을 수도 있다.

: reverse()를 이용한 방법이 조금 더 빠른 runtime을 보여준다.
